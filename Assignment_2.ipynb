### Setup: Preprocessing

```python
df = pd.read_csv('Assignment_2/data_raw/train.csv')
print(df.head())
```

## EDA

```python
# count of missing values
missing_counts = df.isnull().sum()

# calculate percentages
missing_pct = (df.isnull().sum() / len(df)) * 100

# 3. Combine for a clear report
missing_report = pd.DataFrame({
    'Missing Count': missing_counts,
    'Percentage (%)': missing_pct
}).sort_values(by='Percentage (%)', ascending=False)

print(missing_report)
```

Findings:

1) New_Price column value is missing in >86% of records. Filling in such a large percentage of the column with averages, modes, etc from such a small sample would introduce alot of bias and noise. On the other hand, dropping 86% of the rows  leaves almost no data to analyze. For these reasons I think this column should be dropped all together

2) Seats, Engine, Power, and Mileage on the other hand have over 99% non-null values. I think these records can be safely dropped. This allows the model to learn from real data without significantly reducing the size of the original dataset.
